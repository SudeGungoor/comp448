```python
def read_image(imName):
    im = cv.imread(imName)
    return np.flip(im, 2)

def write_image(im, imName):
    im = cv.imwrite(imName, cv.cvtColor(im, cv.COLOR_RGB2BGR))
```

```python
def plot_image(im, w, h, cmap = "gray", title = False):
    fig = plt.figure(figsize = (w, h))
    plt.imshow(im, cmap = cmap, interpolation = "none")
    plt.xticks([])
    plt.yticks([])
    if title:
        plt.title(title)
```

```python
classes = os.listdir(os.path.join(dataset, "Train"))
tr_path = os.path.join(dataset, "Train")
val_path = os.path.join(dataset, "Validation")
ts_path = os.path.join(dataset, "Test")
for c in classes:
  print(f"Class {c} images: Training {len(os.listdir(os.path.join(tr_path, c)))} , Validation {len(os.listdir(os.path.join(val_path, c)))}, Test {len(os.listdir(os.path.join(ts_path, c)))}")
```

```python
def random_flip(image):
  i = np.random.randint(4)
  if i == 0:
    return cv.flip(image, 0)
  elif i == 1:
    return cv.flip(image, 1)
  elif i == 2:
    return cv.flip(cv.flip(image, 0), 1)
  elif i == 3:
    return image
```

```python
def reduce_dataset(input_path, output_path):
  classes = os.listdir(input_path)
  for c in classes:
    print(f"reducing images from {c}")
    os.mkdir(os.path.join(output_path, c))
    images = os.listdir(os.path.join(input_path, c))
    n = len(images)
    if n >= 1000:
      choices = np.random.choice(images, 1000, replace = False)
      for image in choices:
        img = read_image(os.path.join(input_path, c, image))
        write_image(img, os.path.join(output_path, c, image))

    else:
      for image in images:
        img = read_image(os.path.join(input_path, c, image))
        write_image(img, os.path.join(output_path, c, image))
      while n < 500:
        imname = np.random.choice(images, 1)[0]
        img = read_image(os.path.join(input_path, c, imname))
        randomFlipped = random_flip(img)
        write_image(randomFlipped, os.path.join(output_path, c, imname[:-4] + f"{n}" + ".jpg"))
        n += 1
```

```python
os.mkdir(os.path.join(dataset, "Train_reduced"))
reduce_dataset(tr_path, os.path.join(dataset, "Train_reduced"))
```
